```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Smoking Data.csv')
```

```python
df.head()
```

```python
df.drop('Days', axis='columns',inplace=True)
```

```python
df.head()
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1 = df.copy()
```

for col in df.columns:
    if df1[col].dtype == np.object and df1[col] != ['Wear Jacket?']:
        df1[col] = le.fit_transform(df1[col])

```python
df1['Outlook'] = le.fit_transform(df1['Outlook'])
df1['Temprature'] = le.fit_transform(df1['Temprature'])
df1['Routine'] = le.fit_transform(df1['Routine'])
```

```python
df1.head()
```

```python
df.head()
```

```python
x = df1.drop('Wear Jacket?',axis=1)
y = df1['Wear Jacket?']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
clf.fit(x,y)
```

```python
clf.predict([[1,0,0]]) #sunny,cloudy(1,0) ; cold,warm(0,1); indoor,outdoor(0,1)
```

```python
clf.predict([[1,0,1]])
```

# Tree visualise

```python
from sklearn import tree
```

```python
tree.plot_tree(clf)
```

```python
plt.figure(figsize=(12,12)) 
tree.plot_tree(clf)
```

```python
plt.figure(figsize=(12,12)) 
tree.plot_tree(clf,filled=True,rounded=True)
```
